```java
public class Student {
    // Instance variables
    private String name;
    private int score;

    // Default constructor
    public Student() {
        name = "";
        score = 0; // default score is 0
    }

    // Overloaded constructor to initialize name and score
    public Student(String n, int s) {
        name = n;
        setScore(s); // Using setScore to ensure score is within bounds
    }

    // Accessor (getter) methods
    public String getName() {
        return name;
    }

    public int getScore() {
        return score;
    }

    // Mutator (setter) methods
    public void setName(String n) {
        name = n;
    }

    // Set score and ensure it is between 0 and 100
    public void setScore(int s) {
        if (s < 0 || s > 100) {
            score = 0; // Invalid score defaults to 0
        } else {
            score = s;
        }
    }

    // Method to check if the student passes (score >= 70)
    public boolean pass() {
        return score >= 70;
    }
}
public class Main {
    public static void main(String[] args) {
        // Creating a student
        Student student1 = new Student("Alice", 85);
        Student student2 = new Student("Bob", 65);

        // Checking if the students pass
        System.out.println(student1.getName() + " passes: " + student1.pass()); // Alice passes: true
        System.out.println(student2.getName() + " passes: " + student2.pass()); // Bob passes: false
    }
}
Main.main(null)
```

## 5.7 Popcorn Hack #2

```java
import java.util.ArrayList;
import java.util.List;

public class Gadget {
    public static int totalGadgets = 0;  // Static variable to track total gadgets made
    public static List<Gadget> gadgetsList = new ArrayList<>();  // Static list to track all gadgets
    private String gadgetName;  // Instance variable to store the name of the gadget
    private double cost;  // Instance variable to store the cost of the gadget

    // Constructor to set the gadget name, cost, and increment totalGadgets
    public Gadget(String gadgetName, double cost) {
        this.gadgetName = gadgetName;
        this.cost = cost;
        totalGadgets++;  // Increment the total gadgets count
        gadgetsList.add(this);  // Add this gadget to the static list
    }

    // Getter method to retrieve gadget name
    public String getGadgetName() {
        return gadgetName;
    }

    // Getter method to retrieve gadget cost
    public double getCost() {
        return cost;
    }

    // Static method to calculate the total cost of all gadgets
    public static double calculateTotalCost() {
        double totalCost = 0;
        for (Gadget gadget : gadgetsList) {
            totalCost += gadget.getCost();
        }
        return totalCost;
    }
}
public class Main {
    public static void main(String[] args) {
        // Create three gadgets with their respective costs
        Gadget g1 = new Gadget("Freeze Ray", 299.99);
        Gadget g2 = new Gadget("Banana Blaster", 149.99);
        Gadget g3 = new Gadget("Lipstick Taser", 79.99);

        // Print the total number of gadgets
        System.out.println("Total gadgets made: " + Gadget.totalGadgets);

        // Print the list of all created gadgets and their costs
        System.out.println("List of all gadgets created:");
        for (Gadget gadget : Gadget.gadgetsList) {
            System.out.println("- " + gadget.getGadgetName() + ": $" + gadget.getCost());
        }

        // Calculate and print the total cost of all gadgets
        double totalCost = Gadget.calculateTotalCost();
        System.out.printf("Total cost of all gadgets: $%.2f%n", totalCost);
    }
}

Main.main(null);
```

```java
import java.util.ArrayList;
import java.util.List;

public class Minion {
    // Instance variables
    private String color;
    private String name;
    private int energyLevel;
    private List<String> gadgets;  // List to hold gadgets
    private String hair;
    private double height;

    // Default constructor for NPC minions
    public Minion() {
        this.color = "Yellow"; // Default color
        this.name = "Minion";  // Default name
        this.energyLevel = 75;  // Default energy level
        this.hair = "Sprout-Cut"; // Default hair style
        this.height = 3.5; // Default height in feet
        this.gadgets = new ArrayList<>(); // Initialize the gadgets list
        this.gadgets.add("Banana"); // Default gadget
    }

    // Parameterized constructor for main-character minions
    public Minion(String color, String name, int energyLevel, String hair, double height) {
        this.color = color;
        this.name = name;
        this.energyLevel = energyLevel;
        this.hair = hair;
        this.height = height;
        this.gadgets = new ArrayList<>(); // Initialize the gadgets list
    }

    // Method to add a gadget to the minion's collection
    public void addGadget(String gadget) {
        gadgets.add(gadget);
        System.out.println(gadget + " has been added to " + name + "'s gadgets.");
    }

    // Getter methods
    public String getColor() {
        return color;
    }

    public String getName() {
        return name;
    }

    public int getEnergyLevel() {
        return energyLevel;
    }

    public String getHair() {
        return hair;
    }

    public double getHeight() {
        return height;
    }

    public List<String> getGadgets() {
        return gadgets;
    }

    // Method to display the minion's details
    public void displayDetails() {
        System.out.println("Minion Name: " + name);
        System.out.println("Color: " + color);
        System.out.println("Energy Level: " + energyLevel);
        System.out.println("Hair: " + hair);
        System.out.println("Height: " + height + " feet");
        System.out.println("Gadgets: " + gadgets);
    }

    public static void main(String[] args) {
        // Creating default NPC minion
        Minion npcMinion = new Minion();

        // Creating main-character minions using the parameterized constructor
        Minion kevin = new Minion("Yellow", "Kevin", 100, "Sprout-Cut", 4.0);
        Minion bob = new Minion("Purple", "Bob", 50, "None", 3.5);
        Minion gru = new Minion("Grey", "Gru", 200, "Bald", 5.8); // Example of Gru

        // Adding gadgets to specific minions
        kevin.addGadget("Freeze Ray");
        kevin.addGadget("Banana Blaster");
        bob.addGadget("Lipstick Taser");
        gru.addGadget("Evil Minion Ray");

        // Displaying details of all minions
        System.out.println("\n=== NPC Minion Details ===");
        npcMinion.displayDetails();
        
        System.out.println("\n=== Specific Minion Details ===");
        kevin.displayDetails();
        System.out.println();
        bob.displayDetails();
        System.out.println();
        gru.displayDetails();
    }
}
Minion.main(null)
```

```java
import java.util.ArrayList;
import java.util.List;

public class Minion {
    // Instance variables
    private String color;
    private String name;
    private int energyLevel;
    private List<String> gadgets; // List to hold gadgets
    private String hair;
    private double height;

    // Static variables to calculate average energy level
    private static int totalEnergyLevel = 0;  // Total energy levels of all minions
    private static int minionCount = 0;        // Count of all minions created

    // Default constructor for NPC minions
    public Minion() {
        this.color = "Yellow"; // Default color
        this.name = "Minion";  // Default name
        this.energyLevel = 75; // Default energy level
        this.hair = "Sprout-Cut"; // Default hair style
        this.height = 3.5; // Default height in feet
        this.gadgets = new ArrayList<>(); // Initialize the gadgets list
        this.gadgets.add("Banana"); // Default gadget
        updateEnergyStats(energyLevel); // Update energy stats for default minion
    }

    // Parameterized constructor for main-character minions
    public Minion(String color, String name, int energyLevel, String hair, double height) {
        this.color = color;
        this.name = name;
        this.energyLevel = energyLevel;
        this.hair = hair;
        this.height = height;
        this.gadgets = new ArrayList<>(); // Initialize the gadgets list
        updateEnergyStats(energyLevel); // Update energy stats for new minion
    }

    // Method to update total energy level and minion count
    private static void updateEnergyStats(int energyLevel) {
        totalEnergyLevel += energyLevel; // Add to total energy level
        minionCount++; // Increment minion count
    }

    // Method to calculate average energy level
    public static double calculateAverageEnergy() {
        if (minionCount == 0) {
            return 0; // Avoid division by zero
        }
        return (double) totalEnergyLevel / minionCount; // Calculate average
    }

    // Method to add a gadget to the minion's collection
    public void addGadget(String gadget) {
        gadgets.add(gadget);
    }

    // Getter methods
    public String getColor() {
        return color;
    }

    public String getName() {
        return name;
    }

    public int getEnergyLevel() {
        return energyLevel;
    }

    public String getHair() {
        return hair;
    }

    public double getHeight() {
        return height;
    }

    public List<String> getGadgets() {
        return gadgets;
    }

    // Method to display the minion's details
    public void displayDetails() {
        System.out.println("Minion Name: " + name);
        System.out.println("Color: " + color);
        System.out.println("Energy Level: " + energyLevel);
        System.out.println("Hair: " + hair);
        System.out.println("Height: " + height + " feet");
        System.out.println("Gadgets: " + gadgets);
        System.out.println();
    }

    public static void main(String[] args) {
        // Create three minions
        Minion minion1 = new Minion("Yellow", "Kevin", 100, "Sprout-Cut", 4.0);
        Minion minion2 = new Minion("Purple", "Bob", 50, "None", 3.5);
        Minion minion3 = new Minion("Green", "Dave", 80, "Curly", 4.5);

        // Adding gadgets to the minions
        minion1.addGadget("Freeze Ray");
        minion2.addGadget("Banana Blaster");
        minion3.addGadget("Lipstick Taser");

        // Displaying details of all minions
        System.out.println("=== Minion Details ===");
        minion1.displayDetails();
        minion2.displayDetails();
        minion3.displayDetails();

        // Calculate and display average energy level
        double averageEnergy = Minion.calculateAverageEnergy();
        System.out.println("Average Energy Level of all Minions: " + averageEnergy);
    }
}
Minion.main(null)
```

```java
import java.util.ArrayList;
import java.util.List;

public class Minion {
    // Instance variables
    private String color;
    private String name;
    private int energyLevel;
    private List<String> gadgets; // List to hold gadgets
    private String hair;
    private double height;

    // Static variables to calculate average energy level
    private static int totalEnergyLevel = 0;  // Total energy levels of all minions
    private static int minionCount = 0;        // Count of all minions created

    // Default constructor for NPC minions
    public Minion() {
        this.color = "Yellow"; // Default color
        this.name = "Minion";  // Default name
        this.energyLevel = 75; // Default energy level
        this.hair = "Sprout-Cut"; // Default hair style
        this.height = 3.5; // Default height in feet
        this.gadgets = new ArrayList<>(); // Initialize the gadgets list
        this.gadgets.add("Banana"); // Default gadget
        updateEnergyStats(energyLevel); // Update energy stats for default minion
    }

    // Parameterized constructor for main-character minions
    public Minion(String color, String name, int energyLevel, String hair, double height) {
        this.color = color;
        this.name = name;
        this.energyLevel = energyLevel;
        this.hair = hair;
        this.height = height;
        this.gadgets = new ArrayList<>(); // Initialize the gadgets list
        updateEnergyStats(energyLevel); // Update energy stats for new minion
    }

    // Method to update total energy level and minion count
    private static void updateEnergyStats(int energyLevel) {
        totalEnergyLevel += energyLevel; // Add to total energy level
        minionCount++; // Increment minion count
    }

    // Method to calculate average energy level
    public static double calculateAverageEnergy() {
        if (minionCount == 0) {
            return 0; // Avoid division by zero
        }
        return (double) totalEnergyLevel / minionCount; // Calculate average
    }

    // Method to calculate recovery time based on color and energy level
    public int calculateRecoveryTime() {
        int recoveryTime = 0;

        // Check color
        if (this.color.equalsIgnoreCase("purple")) {
            recoveryTime += 10; // Extra recovery time for purple minions
        }

        // Check energy level
        if (this.energyLevel < 50) {
            recoveryTime += 5; // Additional recovery time for low energy
        } else {
            recoveryTime += 3; // Standard recovery time for healthy energy levels
        }

        return recoveryTime; // Return total recovery time
    }

    // Method to add a gadget to the minion's collection
    public void addGadget(String gadget) {
        gadgets.add(gadget);
    }

    // Method to display the minion's details
    public void displayDetails() {
        System.out.println("Minion Name: " + name);
        System.out.println("Color: " + color);
        System.out.println("Energy Level: " + energyLevel);
        System.out.println("Hair: " + hair);
        System.out.println("Height: " + height + " feet");
        System.out.println("Gadgets: " + gadgets);
        System.out.println("Recovery Time: " + calculateRecoveryTime() + " hours");
        System.out.println();
    }

    public static void main(String[] args) {
        // Create three minions
        Minion minion1 = new Minion("Yellow", "Kevin", 100, "Sprout-Cut", 4.0);
        Minion minion2 = new Minion("Purple", "Bob", 40, "None", 3.5);
        Minion minion3 = new Minion("Green", "Dave", 80, "Curly", 4.5);

        // Adding gadgets to the minions
        minion1.addGadget("Freeze Ray");
        minion2.addGadget("Banana Blaster");
        minion3.addGadget("Lipstick Taser");

        // Displaying details of all minions
        System.out.println("=== Minion Details ===");
        minion1.displayDetails();
        minion2.displayDetails();
        minion3.displayDetails();

        // Calculate and display average energy level
        double averageEnergy = Minion.calculateAverageEnergy();
        System.out.println("Average Energy Level of all Minions: " + averageEnergy);
    }
}
Minion.main(null)
```
