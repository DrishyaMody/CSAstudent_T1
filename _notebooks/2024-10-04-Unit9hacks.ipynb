## 9.1/9.2 PopCorn Hack

````java
// Superclass: Shape
public class Shape {
    protected String name;
    protected int sides;

    public Shape(String name, int sides) {
        this.name = name;
        this.sides = sides;
    }

    public void displayInfo() {
        System.out.println("Shape: " + name + ", Sides: " + sides);
    }
}

// Subclass: Circle
public class Circle extends Shape {
    private double radius;

    public Circle(double radius) {
        super("Circle", 0);  // A circle has 0 sides
        this.radius = radius;
    }

    public double area() {
        return Math.PI * radius * radius;
    }

    @Override
    public void displayInfo() {
        super.displayInfo();
        System.out.println("Radius: " + radius + ", Area: " + area());
    }
}

// Subclass: Hexagon
public class Hexagon extends Shape {
    private double sideLength;

    public Hexagon(double sideLength) {
        super("Hexagon", 6);  // A hexagon has 6 sides
        this.sideLength = sideLength;
    }

    public double perimeter() {
        return sideLength * 6;
    }

    @Override
    public void displayInfo() {
        super.displayInfo();
        System.out.println("Side Length: " + sideLength + ", Perimeter: " + perimeter());
    }
}

// Main class to test the implementation
public class Main {
    public static void main(String[] args) {
        Circle circle = new Circle(5);
        circle.displayInfo();

        Hexagon hexagon = new Hexagon(3);
        hexagon.displayInfo();
    }
}
Main.main(null)
````

## 9.3/9.4 Popcorn Hack 2

````java
// Superclass: Shape
public abstract class Shape {
    protected String name;
    protected double length;
    protected double width;

    // Constructor
    public Shape(String name, double length, double width) {
        this.name = name;
        this.length = length;
        this.width = width;
    }

    // Abstract methods to be implemented by subclasses
    public abstract double calc_area();
    public abstract double calc_perimeter();

    // Display info about the shape
    public void displayInfo() {
        System.out.println("Shape: " + name);
    }
}

// Subclass: Triangle
public class Triangle extends Shape {
    private int side1;
    private int side2;
    private int side3;

    // Default constructor
    public Triangle() {
        super("triangle", 0, 0); // Call to Shape constructor with default values
        this.side1 = 1;
        this.side2 = 2;
        this.side3 = 3;
    }

    // Constructor that takes a name and three side lengths
    public Triangle(String name, int s1, int s2, int s3) {
        super(name, 0, 0); // Call to Shape constructor to set the name
        this.side1 = s1;
        this.side2 = s2;
        this.side3 = s3;
    }

    // Override the calc_area method to use Heron's formula
    @Override
    public double calc_area() {
        // Calculate the semi-perimeter
        double s = (side1 + side2 + side3) / 2.0;

        // Apply Heron's formula
        double area = Math.sqrt(s * (s - side1) * (s - side2) * (s - side3));

        return area;
    }

    // Override calc_perimeter method
    @Override
    public double calc_perimeter() {
        // Perimeter is the sum of all three sides
        return side1 + side2 + side3;
    }

    // Override displayInfo method to include triangle details
    @Override
    public void displayInfo() {
        System.out.println("Shape: " + name);
        System.out.println("Sides: " + side1 + ", " + side2 + ", " + side3);
        System.out.println("Perimeter: " + calc_perimeter()); // Show perimeter
        System.out.println("Area (Heron's formula): " + calc_area()); // Show area
    }
}

// Main class to test the Triangle implementation
public class Main {
    public static void main(String[] args) {
        // Create a Triangle object with sides 3, 4, 5
        Triangle triangle = new Triangle("Triangle", 3, 4, 5);
        
        // Display the triangle's details
        triangle.displayInfo();
        
        // Expected output:
        // Shape: Triangle
        // Sides: 3, 4, 5
        // Perimeter: 12.0
        // Area (Heron's formula): 6.0
    }
}
Main.main(null)
````

## 9.6 PopCorn Hack

````java
// Base class: Shape
public abstract class Shape {
    // Abstract method for drawing the shape
    public abstract String draw();
}

// Subclass: Triangle
public class Triangle extends Shape {
    @Override
    public String draw() {
        return "Drawing a triangle.";
    }
}

// Subclass: Rectangle
public class Rectangle extends Shape {
    @Override
    public String draw() {
        return "Drawing a rectangle.";
    }
}

// Subclass: Hexagon
public class Hexagon extends Shape {
    @Override
    public String draw() {
        return "Drawing a hexagon.";
    }
}

// Main class to demonstrate polymorphism
public class Main {
    public static void main(String[] args) {
        // Create an array of Shape references
        Shape[] shapes = new Shape[3];
        
        // Assign different shapes to the array
        shapes[0] = new Triangle();
        shapes[1] = new Rectangle();
        shapes[2] = new Hexagon();
        
        // Iterate through the array and call the draw method
        for (Shape shape : shapes) {
            System.out.println(shape.draw());  // Calls the draw method of the actual object type
        }
    }
}
Main.main(null)
````

# Static Types vs. Dynamic Types in Polymorphism Example

In the provided polymorphism example, static types and dynamic types play key roles in understanding how polymorphism works in Java. Here's a breakdown of each part:

## Static Types
- The **static type** of a variable is the type that is defined at compile time. It is determined by the declaration of the variable.
- In our example, the static types are:
  - `Shape[] shapes` (the array of shape references)
  - `Shape shape` (the variable used in the for-each loop)
  - When we instantiate objects, such as `new Triangle()`, `new Rectangle()`, and `new Hexagon()`, the static type is `Triangle`, `Rectangle`, and `Hexagon`, respectively, but the reference type is `Shape`.

## Dynamic Types
- The **dynamic type** of a variable is the actual type of the object that the variable refers to at runtime. It is determined when the program is executing.
- In our example, the dynamic types are:
  - For `shapes[0]`, the dynamic type is `Triangle`.
  - For `shapes[1]`, the dynamic type is `Rectangle`.
  - For `shapes[2]`, the dynamic type is `Hexagon`.
- When calling `shape.draw()`, the dynamic type determines which `draw()` method implementation is executed at runtime. For instance:
  - If `shape` is referring to a `Triangle` object, then `shape.draw()` will call the `draw()` method of the `Triangle` class.

## Key Points
- **Polymorphism**: It allows us to use the static type (`Shape`) to refer to objects of any subclass type (`Triangle`, `Rectangle`, `Hexagon`). This is what enables polymorphism, as the same method call can result in different behaviors depending on the dynamic type.
- **Method Resolution**: The Java runtime uses the dynamic type to resolve which method to call. This allows subclasses to provide their specific implementations of methods defined in the superclass, while still being treated as instances of the superclass in the code.

## Example Breakdown
Here’s a breakdown of the relevant code sections highlighting static and dynamic types:

```java
// Static Type: Shape[]
Shape[] shapes = new Shape[3];

// Assigning dynamic types to the static type
shapes[0] = new Triangle(); // Static type: Shape, Dynamic type: Triangle
shapes[1] = new Rectangle(); // Static type: Shape, Dynamic type: Rectangle
shapes[2] = new Hexagon(); // Static type: Shape, Dynamic type: Hexagon

// Looping through the shapes array
for (Shape shape : shapes) { // Static type: Shape, Dynamic type depends on the current iteration
    System.out.println(shape.draw()); // Dynamic type determines which draw() method is called
}

# Down-Casting in Java

## Definition
**Down-casting** is the process of converting a reference of a superclass type into a reference of a subclass type. This is useful when you want to access subclass-specific methods or fields that are not present in the superclass. However, down-casting must be done carefully because if the object being cast is not actually an instance of the subclass, a `ClassCastException` will occur at runtime.

## Example of Down-Casting in Polymorphism

Using the previous shape hierarchy, we can demonstrate down-casting as follows:

```java
// Base class: Shape
public abstract class Shape {
    public abstract String draw();
}

// Subclass: Triangle
public class Triangle extends Shape {
    @Override
    public String draw() {
        return "Drawing a triangle.";
    }
}

// Subclass: Rectangle
public class Rectangle extends Shape {
    @Override
    public String draw() {
        return "Drawing a rectangle.";
    }
}

// Subclass: Hexagon
public class Hexagon extends Shape {
    @Override
    public String draw() {
        return "Drawing a hexagon.";
    }

    // Additional method specific to Hexagon
    public String drawWithSides(int sides) {
        return "Drawing a hexagon with " + sides + " sides.";
    }
}

// Main class to demonstrate polymorphism and down-casting
public class Main {
    public static void main(String[] args) {
        Shape[] shapes = new Shape[3];
        shapes[0] = new Triangle();
        shapes[1] = new Rectangle();
        shapes[2] = new Hexagon();

        for (Shape shape : shapes) {
            System.out.println(shape.draw());

            // Down-casting example
            if (shape instanceof Hexagon) { // Check if the shape is a Hexagon
                Hexagon hexagon = (Hexagon) shape; // Down-casting to Hexagon
                System.out.println(hexagon.drawWithSides(6)); // Call the specific method
            }
        }
    }
}


## 9.7 Popcorn Hack

````java
// Base Class: Animal
public class Animal {
    private String name;

    public Animal(String name) {
        this.name = name;
    }

    // Override toString() method to provide meaningful output
    @Override
    public String toString() {
        return name; // Return the name of the animal
    }
}

// Subclass: Dog
public class Dog extends Animal {
    public Dog(String name) {
        super(name);
    }

    // Overriding the toString() method to change the default behavior
    @Override
    public String toString() {
        return "This is a dog named " + super.toString(); // Uses Animal's toString()
    }
}

// Main Class
public class Main {
    public static void main(String[] args) {
        Animal animal = new Animal("Generic Animal");
        Dog dog = new Dog("Buddy");

        // Execute unchanged method from Object (default toString)
        System.out.println("Animal Object: " + animal); // Calls Object's toString()

        // Execute changed method from Object (overridden toString)
        System.out.println("Dog Object: " + dog); // Calls Dog's overridden toString()
    }
}


Main.main(null)
````
